```python
import csv

# booksummaries.txt is tab-separated despite the .txt extension
with open("booksummaries.txt", "r", encoding="utf-8") as book_file:
    book_lines = book_file.readlines()

with open("book.csv", "w", newline="", encoding="utf-8") as csv_book_file:
    writer = csv.writer(csv_book_file)
    writer.writerow(['WikipediaID', 'FreebaseID', 'Title', 'Author', 'Publication Date', 'Genres', 'Summary'])
    for line in book_lines:
        parts = line.strip().split('\t')
        writer.writerow(parts)

print("CSV file 'book.csv' created successfully.")
    
```

```python
import pandas as pd

file_path = "book.csv"
df_book = pd.read_csv(file_path)

df_book
```

```python
df = df_book.drop(columns=['WikipediaID', 'FreebaseID', 'Author', 'Publication Date', 'Genres'], axis=1)

df
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
df['Title+Summary'] = df['Title'] + ', ' + df['Summary']

df_book['embeddings2'] = df['Title+Summary'].apply(lambda x: model.encode(x))

df_book
```

```python
df_book.to_csv("book.csv", index=False)
```

```python
query = "War"

query_embedding = model.encode(query)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def recherche_book(query_embedding, df, top_k=5):

    # Calculate cosine similarity between the query embedding and the embeddings in the DataFrame
    similarities = cosine_similarity([query_embedding], df['embeddings2'].tolist())[0]
    
    # Get the indices of the top_k most similar items
    top_indices = np.argsort(similarities)[-top_k:][::-1]
    
    # Return the top_k most similar items
    return df.iloc[top_indices, 3]

print(recherche_book(query_embedding, df_book, top_k=5))
```

```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer

# Load data
csv_path = "book.csv"
df = pd.read_csv(csv_path)
df = df.drop(columns=['WikipediaID', 'FreebaseID', 'Author', 'Publication Date', 'Genres'], axis=1)

# Prepare texts for embedding
texts = (df['Title'] + ', ' + df['Summary']).tolist()

# Load model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Compute embeddings in batch
embeddings = model.encode(texts, show_progress_bar=True, batch_size=64)

# Save embeddings as .npy
np.save("book_embeddings.npy", embeddings)
print(f"Saved {embeddings.shape[0]} embeddings to RAG/book_embeddings.npy")
```
